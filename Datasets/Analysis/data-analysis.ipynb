## Adult Dataset

```python
df = pd.read_csv("../Cleaned/Adult/Adult.csv", skipinitialspace=True)
df.info()
```

### Classification target distribution

```python
df['target'].value_counts()
```

```python
#adult_df['target'].plot.pie(autopct='%1.1f%%')
df['target'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

### Protected Attributes

#### Sex

```python
df['sex'].value_counts()
```

```python
df['sex'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df.loc[df['sex']=="Male"]['target'].value_counts()
```

```python
df.loc[df['sex']=="Male"]['target'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df.loc[df['sex']=="Female"]['target'].value_counts()
```

```python
df.loc[df['sex']=="Female"]['target'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df_adult_sex = df.groupby(['sex', 'target'])['target'].count()
print(df_adult_sex)

```

```python
df_adult_sex.plot(kind='pie',autopct='%1.1f%%')
```

#### race

```python
df['race'].value_counts()
```

```python
df['race'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df_adult_race = df.groupby(['race', 'target'])['target'].count()
print(df_adult_race)
```

```python
#adult_race.plot(kind='pie',autopct='%1.1f%%')
df_adult_race.plot(kind='bar')
```
